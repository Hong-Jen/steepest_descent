```python
from sympy import *

def equation(x):
    return x**3+x**2-10*x+25

x = Symbol('x')

def derivative(x):
    return diff(equation(x),x)

#print(derivative(x))

r = [0.1,0.05,0.01,0.005,0.001,0.0005] #Assumption
min_tol = 0.005 #minimum tolerance range
max_tol = 1000 #maximum tolerance range
m_step = 2000 #maximum step

# print("Initial guess = ",cur_x)

for j in range (0,len(r)):
    print("##############################")
    print('r=',r[j])
    cur_x = [-2,10]  #Initial point
    length = len(cur_x) #numbers of Initial point 
    for i in range (0,length):
        step = 0
        condition = True
        print('Initial point=',cur_x[i])
        x_path=[cur_x[i]]
        while condition:
            n_x = cur_x[i]
            n_x = n_x - r[j]*(derivative(x).subs(x,n_x))
            error = abs(n_x - cur_x[i])
    #         print('----------------------------')
    #         print("Step", step, "\nX"+str(step+1), " = ", n_x)
    #         print('error = '+ str(error))

            step = step + 1
            if step > m_step:
                print(cur_x[i],"is not convergence in given maximum step")
                break
            x_path.append(n_x)
            cur_x[i]= n_x
            condition = error > min_tol and error < max_tol
        if error>min_tol:
            print('This equation is divergent')
#         print(x_path)
        print('step=',step-1)
        print('error = '+ str(error))
        print('final point=',cur_x[i])
        print('----------------------------')

```

```python
from sympy import *
def equation(x1,x2):
    return (11-x1-x2)**2+(1+10*x2+x1-x1*x2)**2

x1 = Symbol('x1')
x2 = Symbol('x2')

def x1_derivative(x1, x2):
    return diff(equation(x1,x2),x1)
def x2_derivative(x1, x2):
    return diff(equation(x1,x2),x2)


cur_x1 = [10] #Initial point for x1
cur_x2 = [15]  #Initial point for x2
r = 0.005 #Assumption
min_tol = 0.0001 #minimum tolerance range
max_tol = 10000 #maximum tolerance range
m_step = 2000 #maximum step

# print("Initial guess = ",cur_x)

length = len(cur_x1) #numbers of Initial point 

for i in range (0, length):
    step = 0
    condition = True
    print('\nInitial point=',[cur_x1[i], cur_x2[i]])
    while condition:
        n_x1 = cur_x1[i]
        n_x2 = cur_x2[i]
#         x1_path=np.linspace(cur_x1[i])
#         x2_path=np.linspace(cur_x2[i])
        a = x1_derivative(x1, x2).subs(x1, cur_x1[i]).subs(x2, cur_x2[i])
        b = x2_derivative(x1, x2).subs(x1, cur_x1[i]).subs(x2, cur_x2[i])
        nor_const = (a**2+b**2)**(1/2)
        n_x1 = n_x1 - r*(a)/nor_const
        n_x2 = n_x2 - r*(b)/nor_const
        
        x1_error = abs(n_x1 - cur_x1[i])
        x2_error = abs(n_x2 - cur_x2[i])
#         print('----------------------------')
#         print("Step", step, "\nX"+str(step+1), " = ", n_x)
#         print('error = '+ str(error))

        step = step + 1
        if step > m_step:
            print(" [x1, x2] = ", cur_x1[i], cur_x2[i],"is not convergence in given maximum step")
            break

        cur_x1[i] = n_x1
        cur_x2[i] = n_x2
        condition = x1_error > min_tol and x2_error > min_tol and x1_error < max_tol and x2_error < max_tol
    print('r=',r)
    print('step=',step-1)
    print('x1 error = '+ str(x1_error))
    print('x2 error = '+ str(x2_error))
    print('final point [x1, x2] =',[cur_x1[i], cur_x2[i]])
    print('----------------------------')
```
